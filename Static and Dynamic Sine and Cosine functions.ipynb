```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import ipywidgets as widgets
from ipywidgets import interact, interactive, fixed,interact_manual
import warnings
warnings.filterwarnings('ignore')
%config InlineBackend.figure_format = 'svg'
%matplotlib inline
```

## Statis Sine and Cosie wave

```python
#Lets generate x and y
x=np.arange(0,2*np.pi,0.1) #arange defines, start, end and increment
y=np.sin(x)
plt.plot(x,y)
y2=np.cos(x)
plt.plot(x,y2)
```

## Plot them with labels

```python
x=np.arange(0,4*np.pi,0.1) #range defines, start, end and increment
y=np.sin(x)
plt.plot(x,y)
y2=np.cos(x)
plt.plot(x,y2)
##Labels block
plt.xlabel('x values from 0 to 4pi')  # string must be enclosed with quotes '  '
plt.ylabel('sin(x) and cos(x)')
plt.title('Plot of sin and cos from 0 to 4pi')
plt.legend(['sin(x)', 'cos(x)'])      # legend entries as seperate strings in a list
```

## Plot Dynamic Sine and Cosine Curves

```python
def plot_func(freq):
    x = np.linspace(0,2*np.pi)
    y = np.sin(x * freq)
    plt.plot(x, y)

interact(plot_func, freq = widgets.FloatSlider(value=7.5,
                                               min=1,
                                               max=5.0,
                                               step=0.5))
```
